# Visualization Packages

```python
import matplotlib.pyplot as plt
```

# Barplot

### 1)
 - x : categorical
 - y : numeric
 - data type : list

```python
x = ['category1', 'cat2', 'cat3', 'cat4']
y = [10, 7, 8, 5]
plt.bar(x, y)
plt.title('title')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['category1', 'cat2', 'cat3', 'cat4']
y = [10, 7, 8, 5]
plt.bar(x, y, width=0.8, bottom=None, align='center')
plt.title('title')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.show()
```

# Pie chart

### (1)
 - labels_li : list or tuple
 - values_li : list or tuple

```python
labels_li = ['component_1', 'comp_2', 'comp_3', 'comp_4']
values_li = [1052, 2245, 340, 777]
explode_li = [0,0,0,0.1]
fig1, ax1 = plt.subplots()
ax1.pie(values_li, labels=labels_li,)
plt.title('title')
plt.show()
```

```python
labels_li = ['component_1', 'comp_2', 'comp_3', 'comp_4']
values_li = [1052, 2245, 340, 777]
explode_li = [0,0,0,0.1]

fig1, ax1 = plt.subplots()

ax1.pie(values_li,
        explode=explode_li,
        labels=labels_li,
        autopct='%1.2f%%',
        shadow=True,
        startangle=90)
# ax1.axis('eqaul')
plt.title('title')
plt.show()
```
