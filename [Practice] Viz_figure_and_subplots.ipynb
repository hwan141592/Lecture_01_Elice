```python
import matplotlib.pyplot as plt

fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

x = range(0,100)      
y = [v*v for v in x]

ax1.plot(x,y)
ax2.bar(x,y)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 2 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)

fig = plt.figure(figsize=(12,3))   # figure size

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.plot(x, sin_y, 'b--')
ax2.plot(x, cos_y, 'r--')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()       # 1 by 1
ax.set_title('A single plot')
ax.plot(np.random.rand(100))
```

```python
fig, axs = plt.subplots(2)        # 2 by 1

axs[0].plot(np.random.rand(5))
axs[1].plot(np.random.rand(100))

#fig.suptitle('Vertically stacked subplots')
```

```python
fig, (kdh, goodcoach) = plt.subplots(2)  # a few Axes

kdh.plot(np.random.rand(100))
goodcoach.plot(np.random.rand(100))

#fig.suptitle('Vertically stacked subplots')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)   # 1 by 2
fig.suptitle('Horizontally stacked subplots')

ax1.plot(np.random.rand(20))
ax2.plot(np.random.rand(20))
```

```python
fig, axs = plt.subplots(2, 2)   # 2 by 2

axs[0, 0].plot(np.random.rand(20))
axs[0, 0].set_title('Axis [0, 0]')

axs[0, 1].plot(np.random.rand(20))
axs[0, 1].set_title('Axis [0, 1]')

axs[1, 0].plot(np.random.rand(20))
axs[1, 0].set_title('Axis [1, 0]')

axs[1, 1].plot(np.random.rand(20))
axs[1, 1].set_title('Axis [1, 1]')


for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# * ndarray.flat : a 1-D iterator over the array.
## >> This is a numpy.flatiter instance, which acts similarly to,
##    but is not a subclass of, Python’s built-in iterator object.
## more : https://stackoverflow.com/questions/46862861/what-does-axes-flat-in-matplotlib-do

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(np.random.rand(20), 'tab:orange')
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].plot(np.random.rand(20), 'tab:green')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].plot(np.random.rand(20), 'tab:blue')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].plot(np.random.rand(20), 'tab:red')
axs[1, 1].set_title('Axis [1, 1]')


# 제외
for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# 제외
for ax in fig.get_axes():
    ax.label_outer()

plt.show()
```

```python
fig, axs = plt.subplots(3, sharex=True, sharey=True)
fig.suptitle('Sharing both axes')
axs[0].plot(np.random.rand(20), 'tab:blue')
axs[1].plot(np.random.rand(20), 'tab:blue')
axs[2].plot(np.random.rand(20), 'tab:blue')
```

## 해석해보기

```python
size_graph = 3

for idx_x in range(size_graph):
    fig, axes = plt.subplots(1, 2, figsize=(25, 3))
    
    axes[0].plot(np.random.rand(100))
    axes[0].set_ylim(0, 1.2)
    axes[0].set_ylabel('Probability', fontsize=30)
    # axes[0].grid()
    
    axes[1].plot(np.random.rand(100))
    axes[1].set_ylim(0, 2.4)
    axes[1].set_ylabel('Probability', fontsize=20)
    axes[1].grid()

plt.show()
```
